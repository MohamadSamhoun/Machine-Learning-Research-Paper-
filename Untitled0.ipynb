```python
import cv2
import numpy as np

def weather_preprocess(img_bgr):
    """
    MW: Weather-aware preprocessing
    - Contrast Limited Adaptive Histogram Equalization (CLAHE)
    - Light dehazing via histogram stretching
    - Gamma correction for illumination
    """
    # 1) Convert to LAB and apply CLAHE on L channel
    lab = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    l_clahe = clahe.apply(l)
    lab_clahe = cv2.merge((l_clahe, a, b))
    img_clahe = cv2.cvtColor(lab_clahe, cv2.COLOR_LAB2BGR)

    # 2) Simple â€œdehazeâ€ via contrast stretching
    img_float = img_clahe.astype(np.float32) / 255.0
    min_val = img_float.min()
    max_val = img_float.max()
    img_stretch = (img_float - min_val) / (max_val - min_val + 1e-6)

    # 3) Gamma correction (brighten dark scenes)
    gamma = 1.2  # < 1 darker, >1 brighter
    img_gamma = np.power(img_stretch, 1.0 / gamma)

    # Back to uint8
    img_out = np.clip(img_gamma * 255.0, 0, 255).astype(np.uint8)
    return img_out
```

```python
import numpy as np

def mc_tta_uncertainty(model, image_path, K=10, conf=0.25, iou=0.5):
    """
    Monte Carlo Test-Time Augmentation:
      - Run model K times with augment=True
      - Return image-level epistemic uncertainty u_image
        as variance of mean confidences across runs
    """
    conf_lists = []

    for k in range(K):
        results = model.predict(
            source=image_path,
            imgsz=640,
            conf=conf,
            iou=iou,
            augment=True,   # stochastic TTA
            verbose=False
        )
        boxes = results[0].boxes
        if boxes is None or len(boxes) == 0:
            conf_lists.append(np.array([]))
        else:
            conf_lists.append(boxes.conf.cpu().numpy())

    mean_confs = [c.mean() for c in conf_lists if c.size > 0]

    if len(mean_confs) == 0:
        # No detections â†’ very uncertain scene
        u_image = 1.0
    else:
        u_image = float(np.var(mean_confs))  # epistemic uncertainty

    return u_image, conf_lists


def compute_risk_for_image(model, image_path, K=10, conf=0.25, iou=0.5,
                           alpha=1.0, beta=1.0):
    """
    Returns:
      u_image: epistemic uncertainty (scalar for the whole image)
      risks: list of dicts, one per detection:
             {bbox, cls, conf, sigma, u_image, risk}
    """
    u_image, conf_runs = mc_tta_uncertainty(model, image_path, K, conf, iou)

    # Single standard prediction (no augment) for final boxes
    result = model.predict(
        source=image_path,
        imgsz=640,
        conf=conf,
        iou=iou,
        verbose=False
    )[0]

    boxes = result.boxes
    if boxes is None or len(boxes) == 0:
        return u_image, []

    confs = boxes.conf.cpu().numpy()
    cls_ids = boxes.cls.cpu().numpy().astype(int)
    xyxy = boxes.xyxy.cpu().numpy()

    risks = []
    for i in range(len(confs)):
        p_i = confs[i]
        sigma_i = 1.0 - p_i          # aleatoric proxy: low conf â†’ high sigma
        r_i = alpha * u_image + beta * sigma_i

        risks.append({
            "bbox": xyxy[i].tolist(),
            "cls": int(cls_ids[i]),
            "conf": float(p_i),
            "sigma": float(sigma_i),
            "u_image": float(u_image),
            "risk": float(r_i)
        })

    return u_image, risks
```

```python
import os
import numpy as np

valid_dir = "/content/datasets/adverse-weather-1/valid/images"
all_imgs = os.listdir(valid_dir)

weather_groups = {
    "dusttornado": [],
    "rain_storm": [],
    "haze": [],
    "mist": [],
    "sand_storm": []
}

# Group file paths by weather keyword in filename
for name in all_imgs:
    full_path = os.path.join(valid_dir, name)
    for key in weather_groups.keys():
        if key in name:
            weather_groups[key].append(full_path)

for k, v in weather_groups.items():
    print(k, ":", len(v), "images")
```

```python
summary = {}

for weather, paths in weather_groups.items():
    if len(paths) == 0:
        continue

    u_values = []
    mean_risks = []

    # Use at most 10 images per weather to keep it fast
    for img_path in paths[:10]:
        u_img, risks = compute_risk_for_image(best_model, img_path, K=10)
        if len(risks) == 0:
            continue

        u_values.append(u_img)
        # average risk across detections in that image
        img_mean_risk = np.mean([r["risk"] for r in risks])
        mean_risks.append(img_mean_risk)

    if len(u_values) > 0:
        summary[weather] = {
            "mean_u": float(np.mean(u_values)),
            "mean_risk": float(np.mean(mean_risks)),
            "num_images": len(u_values)
        }

summary
```

```python
def load_yolo_labels(label_path):
    boxes = []
    with open(label_path, "r") as f:
        for line in f.readlines():
            parts = line.strip().split()
            if len(parts) != 5:
                continue
            cls = int(parts[0])
            x, y, w, h = map(float, parts[1:])
            boxes.append((cls, x, y, w, h))
    return boxes

def yolo_to_xyxy(box, img_w, img_h):
    cls, x, y, w, h = box
    cx = x * img_w
    cy = y * img_h
    bw = w * img_w
    bh = h * img_h
    x1 = int(cx - bw / 2)
    y1 = int(cy - bh / 2)
    x2 = int(cx + bw / 2)
    y2 = int(cy + bh / 2)
    return cls, x1, y1, x2, y2

def explanation_overlap(heatmap, bbox):
    H, W = heatmap.shape
    x1, y1, x2, y2 = bbox
    x1 = np.clip(x1, 0, W-1)
    x2 = np.clip(x2, 0, W-1)
    y1 = np.clip(y1, 0, H-1)
    y2 = np.clip(y2, 0, H-1)

    total = heatmap.sum() + 1e-6
    inside = heatmap[y1:y2, x1:x2].sum()
    return float(inside / total)
```
